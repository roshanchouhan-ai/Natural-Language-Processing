## Performing lemmatization using pos_tag

````python
import nltk
nltk.download("wordnet")
nltk.download("punkt_tab")
nltk.download("averaged_perceptron_tagger_eng")


from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

from nltk.tokenize import word_tokenize
from nltk import pos_tag
````

````python
text = "The children were running quickly towards the beautiful gardens while the dogs chased the frightened cats."

#text = "The book was on the table, so I booked another book online."

words = word_tokenize(text)
print(words)
````

````python
pos_tags = pos_tag(words)
print(pos_tags)
````

````python
def get_wordnet_pos(nltk_tags):
    wordnet_tags=[]
    for word, tag in nltk_tags:
        if tag.startswith("J"):
            wordnet_tags.append((word,wordnet.ADJ))
        elif tag.startswith("V"):
            wordnet_tags.append((word,wordnet.VERB))
        elif tag.startswith("R"):
            wordnet_tags.append((word,wordnet.ADV))
        else:
            wordnet_tags.append((word,wordnet.NOUN))

    return wordnet_tags

wordnet_pos_tags = get_wordnet_pos(pos_tags)
print(wordnet_pos_tags)
````

````python
lemmatizer = WordNetLemmatizer()
lemmatized_words = []

for word, pos_t in wordnet_pos_tags:
    
    lemmatized_words.append(lemmatizer.lemmatize(word, pos=pos_t))

print(lemmatized_words)
````

# Lemmatization

---

# Overview

**Lemmatization** is the process of reducing a word to its **lemma**, i.e., its **dictionary (base) form**, by considering both its **meaning** and its **grammatical role (Part of Speech)**.

Unlike stemming, lemmatization does **not** simply remove prefixes or suffixes. Instead, it uses a **lexical database (WordNet)** and **POS information** to determine the correct base form of a word.

> **Note:** The output of lemmatization is always intended to be a valid dictionary word.

---

# Why is Lemmatization Needed?

Consider the following words:

```text
play
playing
played
plays
```

Although these words represent the same concept, a computer treats them as four different words.

Without Lemmatization:

```python
["play", "playing", "played", "plays"]
```

After Lemmatization:

```python
["play", "play", "play", "play"]
```

Now all these words are treated as the same feature.

---

# Another Example

| Original Word | Lemma |
|---------------|-------|
| studies | study |
| playing | play |
| running | run |
| mice | mouse |
| geese | goose |
| wolves | wolf |
| children | child |
| better (Adjective) | good |

Notice that every lemma is a **valid English dictionary word**.

---

# What is a Lemma?

A **lemma** is the canonical or dictionary form of a word.

Examples:

| Word | Lemma |
|------|-------|
| playing | play |
| played | play |
| studies | study |
| running | run |
| children | child |
| mice | mouse |
| better | good |

Unlike stems, lemmas are meaningful English words.

---

# How Does Lemmatization Work?

Lemmatization does **not** remove suffixes blindly.

Instead, it:

1. Looks up the word in a **lexical database (WordNet)**.
2. Uses the **Part of Speech (POS)** of the word.
3. Returns the correct dictionary form.

For example:

```text
playing
```

↓

WordNet + POS (Verb)

↓

```text
play
```

Another example:

```text
better
```

↓

WordNet + POS (Adjective)

↓

```text
good
```

---

# WordNet

**WordNet** is a large lexical database of the English language.

It stores information about:

- Words
- Meanings
- Synonyms
- Antonyms
- Hypernyms
- Hyponyms
- Lemmas

The `WordNetLemmatizer` uses WordNet internally to find the correct lemma.

---

# Why Does POS Matter?

The same word can perform different grammatical roles.

Example:

```text
Book the tickets.
```

Here,

```text
book
```

is a **Verb**.

---

```text
This book is interesting.
```

Here,

```text
book
```

is a **Noun**.

Without knowing the grammatical role, the lemmatizer cannot always determine the correct base form.

---

# Default Behavior

By default, `WordNetLemmatizer` assumes every word is a **noun**.

Example:

```python
lemmatizer.lemmatize("playing")
```

Output

```text
playing
```

Because "playing" is assumed to be a noun.

---

When specifying the POS:

```python
lemmatizer.lemmatize("playing", pos=wordnet.VERB)
```

Output

```text
play
```

---

# Supported WordNet POS Tags

| WordNet POS | Meaning |
|--------------|---------|
| wordnet.NOUN | Noun |
| wordnet.VERB | Verb |
| wordnet.ADJ | Adjective |
| wordnet.ADV | Adverb |

# Advantages

- Produces valid dictionary words
- Uses grammatical information
- More accurate than stemming
- Handles irregular word forms
- Preserves the meaning of words

---

# Disadvantages

- Slower than stemming
- Requires WordNet
- Often requires POS tagging
- Computationally more expensive

---

# Stemming vs Lemmatization

| Feature | Stemming | Lemmatization |
|----------|-----------|---------------|
| Uses Rules | ✅ Yes | ❌ No |
| Uses Dictionary | ❌ No | ✅ Yes (WordNet) |
| Uses POS Information | ❌ No | ✅ Yes |
| Produces Dictionary Words | ❌ Not Always | ✅ Yes |
| Speed | Faster | Slower |
| Accuracy | Lower | Higher |

Example:

| Original | Stem | Lemma |
|-----------|------|--------|
| studies | studi | study |
| happiness | happi | happiness |
| better | better / bet | good |
| children | children | child |

# Complete Example: Tokenization → POS Tagging → Lemmatization

```python
import nltk

nltk.download("punkt_tab")
nltk.download("wordnet")
nltk.download("averaged_perceptron_tagger_eng")

from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

text = "The children were running quickly towards the beautiful gardens while the dogs chased the frightened cats."

# Tokenization
tokens = word_tokenize(text)

# POS Tagging
pos_tags = pos_tag(tokens)

# Convert Penn Treebank Tags to WordNet Tags
def get_wordnet_pos(tag):

    if tag.startswith("J"):
        return wordnet.ADJ

    elif tag.startswith("V"):
        return wordnet.VERB

    elif tag.startswith("N"):
        return wordnet.NOUN

    elif tag.startswith("R"):
        return wordnet.ADV

    else:
        return wordnet.NOUN


lemmatizer = WordNetLemmatizer()

lemmatized_words = []

for word, tag in pos_tags:
    wn_tag = get_wordnet_pos(tag)
    lemma = lemmatizer.lemmatize(word, pos=wn_tag)
    lemmatized_words.append(lemma)

print("Original Tokens:")
print(tokens)

print("\nPOS Tags:")
print(pos_tags)

print("\nLemmatized Words:")
print(lemmatized_words)
```

# Key Takeaways

- **Lemmatization** reduces words to their **dictionary (base) form**, called a **lemma**.
- It uses **WordNet** and **Part-of-Speech (POS)** information instead of simply removing suffixes.
- Unlike stemming, the output is intended to be a **valid English word**.
- By default, `WordNetLemmatizer` assumes every word is a **noun**, so providing the correct POS significantly improves accuracy.
- In real-world NLP pipelines, **POS Tagging is usually performed before Lemmatization**.
- Lemmatization is widely used in text preprocessing, search engines, information retrieval, text classification, and many classical NLP applications where preserving the actual meaning of words is important.

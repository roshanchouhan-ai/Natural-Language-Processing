## Regular Expressions (Regex) Tutorial for NLP

#### What is Regex?

Regex (Regular Expression) is a pattern-matching technique used to search, extract, replace, and validate text.

Think of Regex as a powerful search tool for text.

Examples:

- Extract emails
- Remove special characters
- Find phone numbers
- Detect URLs
- Clean text before NLP

---

## Why Regex is Important in NLP

Before converting text into features (BoW, TF-IDF, Embeddings), we often clean the text.

Regex helps us:

- Remove punctuation
- Remove numbers
- Remove URLs
- Remove HTML tags
- Remove extra spaces
- Extract useful information

Example:

Raw Text:

"Visit https://google.com !!! Call me at 9876543210"

After Regex Cleaning:

"Visit Call me at"

## Basic Regex Functions

#### 1. re.findall()

Returns all matches.

```python

```

Output:


Replace matched patterns.

```python

text = "Hello123"
```

Output:


#### 4. re.split()

Split text using regex pattern.

```python

result = re.split(r"[,;]", text)

```

Output:

```python
['apple', 'banana', 'orange']
```

# Greedy Matching and Backtracking in Regular Expressions

---

# What is Greedy Matching?

By default, most regex quantifiers are **greedy**.

A greedy quantifier tries to match **as many characters as possible** while still allowing the entire regex to match.

The greedy quantifiers are:

| Quantifier | Meaning |
|------------|---------|
| `*` | 0 or more |
| `+` | 1 or more |
| `?` | 0 or 1 (greedy by default) |
| `{n,m}` | Between `n` and `m` times |

---

## Example 1

Text:

```text
aaaaab
```

Regex:

```regex
a+b
```

### Step 1

`a+` tries to consume as many `a`s as possible.

```
aaaaab
^^^^^
```

Remaining:

```
b
```

Regex still needs:

```regex
b
```

Success.

Final Match:

```text
aaaaab
```

---

## Example 2

Text:

```text
123456789
```

Regex:

```regex
\d+
```

`\d+` consumes every digit because nothing follows it.

Match:

```text
123456789
```

---

# What is Backtracking?

Backtracking occurs **when a greedy quantifier has consumed too much text**, preventing the rest of the regex from matching.

The regex engine then:

1. Gives back one character.
2. Tries again.
3. Repeats until the entire regex matches or no solution exists.

---

## Example 3

Text:

```text
aaaaab
```

Regex:

```regex
a+ab
```

### Step 1 (Greedy)

`a+` consumes:

```
aaaaa
^^^^^
```

Remaining:

```
b
```

Regex still needs:

```text
ab
```

Current text:

```text
b
```

Fail.

---

### Step 2 (Backtrack)

Regex gives one `a` back.

Now:

```
aaaa
^^^^
```

Remaining:

```
ab
```

Now the remaining regex:

```regex
ab
```

matches perfectly.

Final Match:

```text
aaaaab
```

---

## Visual Representation

Initial:

```
a+ = aaaaa
Need = ab

FAIL
```

Backtrack:

```
a+ = aaaa
Need = ab

SUCCESS
```

---

# Example 4

Text:

```text
company.co.in
```

Regex:

```regex
[A-Za-z0-9.]+\.[A-Za-z]{2}
```

### Step 1

Greedy part consumes everything.

```
company.co.in
^^^^^^^^^^^^^
```

Nothing remains.

Regex still needs:

```regex
\.[A-Za-z]{2}
```

Fail.

---

### Step 2

Regex backtracks.

Now greedy match becomes:

```
company.co
^^^^^^^^^^
```

Remaining:

```
.in
```

Now:

```regex
\.[A-Za-z]{2}
```

matches:

```
.in
```

Success.

---

# Why Backtracking Exists

Without backtracking, many useful regex patterns would never work.

Consider:

```regex
[A-Za-z0-9.]+\.[A-Za-z]{2,}
```

Without backtracking,

```
company.co.in
```

would never match because the greedy part would consume the entire string.

---

# Example 5

Text:

```html
<div>Hello</div>
```

Regex:

```regex
<.*>
```

`.` means any character.

`*` means zero or more.

Since `*` is greedy,

it matches:

```html
<div>Hello</div>
```

Everything.

---

Now consider:

```html
<div>Hello</div><p>World</p>
```

Regex:

```regex
<.*>
```

Result:

```html
<div>Hello</div><p>World</p>
```

Entire string matched.

This is usually **not** what we want.

---

# Lazy Matching

Adding `?` makes the quantifier **lazy**.

```regex
<.*?>
```

Now regex matches as little as possible.

Example:

```html
<div>Hello</div><p>World</p>
```

Matches:

```
<div>
```

Next match:

```
</div>
```

Next:

```
<p>
```

Next:

```
</p>
```

---

# Greedy vs Lazy

Text:

```html
<div>Hello</div><p>World</p>
```

Greedy:

```regex
<.*>
```

Output:

```html
<div>Hello</div><p>World</p>
```

Lazy:

```regex
<.*?>
```

Output:

```text
<div>
</div>
<p>
</p>
```



Remaining:


Success.

Regex:

### Step 1

`\d+` consumes:

# Catastrophic Backtracking

Some regex patterns create enormous amounts of backtracking.

Example:

```regex
(a+)+
```

Text:

```text
aaaaaaaaaaaaaaaaaaaaX
```

The regex engine tries countless combinations before giving up.

Time complexity can become **exponential**.

These patterns should generally be avoided.

# Summary

## Greedy Quantifiers

| Symbol | Meaning |
|---------|---------|
| `*` | Zero or more (greedy) |
| `+` | One or more (greedy) |
| `?` | Zero or one (greedy) |
| `{n,m}` | Between n and m times (greedy) |

---

## Lazy Quantifiers

Add `?`

| Greedy | Lazy |
|---------|------|
| `*` | `*?` |
| `+` | `+?` |
| `??` | Lazy optional |
| `{n,m}?` | Lazy bounded repetition |

---

# Greedy Matching Algorithm

When regex encounters:

```regex
+
*
{n,m}
```

it follows this strategy:

1. Match as much text as possible.
2. Check whether the remaining regex matches.
3. If not,
   - give one character back,
   - try again.
4. Repeat until:
   - the whole regex succeeds, or
   - no characters remain.

This process is called **Backtracking**.

---

# One-Line Memory Trick

> **Greedy quantifiers take as much as they can. Backtracking gives characters back until the rest of the regex can match.**

---

# Interview Tip

If someone asks:

**"Why does `[A-Za-z0-9.]+\.[A-Za-z]{2,}` correctly match `company.co.in` even though `+` is greedy?"**

The correct answer is:

> The `+` quantifier is greedy and initially consumes the entire string. Since the remaining pattern (`\.[A-Za-z]{2,}`) cannot match, the regex engine backtracks character by character until the final part of the pattern can successfully match `.in`.

# Word Boundary (`\b`), Capturing Groups `()`, and Non-Capturing Groups `(?:)` in Regex

---

# 1. Word Boundary (`\b`)

## What is `\b`?

`\b` means **Word Boundary**.

It does **not** match a character.

Instead, it matches a **position** between:

- a word character (`\w`)
- and a non-word character (`\W`)

or between the start/end of a string and a word character.

Think of it as the **beginning or end of a word**.

---

## What is a Word Character?

`\w` includes:

- `A-Z`
- `a-z`
- `0-9`
- `_`

Everything else is a non-word character.

Examples:

```
Space
.
,
@
#
!
(
)
-
```

---

## Example 1

Text:

```text
cat scatter category cat
```

Regex:

```regex
cat
```

Python:

```python
import re

text = "cat scatter category cat"

re.findall(r"cat", text)
```

Output:

```python
['cat', 'cat', 'cat', 'cat']
```

Because regex matches **every occurrence** of `cat`, even inside larger words.

---

## Example 2

Regex:

```regex
\bcat\b
```

Python:

```python
re.findall(r"\bcat\b", text)
```

Output:

```python
['cat', 'cat']
```

Now regex only matches **complete words**.

---

## Visual Representation

```
cat scatter category cat

|cat| scatter category |cat|
```

Only the first and last `cat` have word boundaries on both sides.

---

## Example 3

Text:

```text
I love NLP
```

Regex:

```regex
\b[A-Z]\w*\b
```

Output:

```python
['I', 'NLP']
```

Explanation:

```
\b      Start of word
[A-Z]   Capital letter
\w*     Remaining characters
\b      End of word
```

---

## Example 4

Extract complete numbers.

Text:

```text
Marks: 95 and 100
```

Regex:

```regex
\b\d+\b
```

Output:

```python
['95', '100']
```

---

# Summary of `\b`

| Regex | Meaning |
|--------|---------|
| `\bcat\b` | Whole word "cat" |
| `\b\d+\b` | Whole numbers |
| `\b\w+\b` | Whole words |
| `\b[A-Z]\w*\b` | Words starting with a capital letter |

# 2. Capturing Groups `()`

## What is a Capturing Group?

Parentheses `()` create a **capturing group**.

A capturing group:

- groups part of a regex
- remembers what it matched
- affects the output of `re.findall()`

---

## Example 1

Text:

```text
9:30 AM
12:45 PM
```

Regex:

```regex
\d{1,2}:\d{2} (AM|PM)
```

Python:

```python
re.findall(r"\d{1,2}:\d{2} (AM|PM)", text)
```

Output:

```python
['AM', 'PM']
```

Notice that `findall()` returns only the captured group.

---

## Example 2

Text:

```text
Username: john123
Username: alice007
```

Regex:

```regex
Username: (\S+)
```

Output:

```python
['john123', 'alice007']
```

Only the usernames are returned because they are inside the capturing group.

---

## Example 3

Without capturing:

Regex:

```regex
Username: \S+
```

Output:

```python
['Username: john123',
 'Username: alice007']
```

Entire match is returned.

---

# Why Use Capturing Groups?

They allow us to extract only the important part of a match.

Instead of:

```
Username: john123
```

we get:

```
john123
```

---

# 3. Non-Capturing Groups `(?: )`

## What is a Non-Capturing Group?

Sometimes we want to group alternatives together, but we **do not** want regex to capture them.

For that we use:

```regex
(?:...)
```

---

## Example 1

Text:

```text
9:30 AM
12:45 PM
```

Regex:

```regex
\d{1,2}:\d{2} (?:AM|PM)
```

Python:

```python
re.findall(r"\d{1,2}:\d{2} (?:AM|PM)", text)
```

Output:

```python
['9:30 AM', '12:45 PM']
```

Unlike capturing groups, the **entire match** is returned.

---

## Compare

### Capturing Group

Regex:

```regex
(AM|PM)
```

Output:

```python
['AM', 'PM']
```

---

### Non-Capturing Group

Regex:

```regex
(?:AM|PM)
```

Output:

```python
['9:30 AM', '12:45 PM']
```

---

# Example 2

Validate Indian mobile numbers.

Regex:

```regex
(?:\+91[- ]?)?[6-9]\d{9}
```

Explanation:

```
(?:...)      Group
\+91         Country code
[- ]?        Optional '-' or space
?            Entire country code is optional
```

Matches:

```
9876543210
+919876543210
+91-9876543210
+91 9876543210
```

---

# Why Not Use `()` Here?

If we wrote:

```regex
(\+91[- ]?)?[6-9]\d{9}
```

then:

```python
re.findall(...)
```

would return:

```python
['+91-', '']
```

instead of the phone numbers.

That's usually not what we want.

---

# Comparison

| Feature | `()` | `(?:)` |
|---------|------|---------|
| Groups expressions | ✅ | ✅ |
| Captures matched text | ✅ | ❌ |
| Affects `findall()` output | ✅ | ❌ |
| Used for extraction | ✅ | ❌ |
| Used only for grouping | ❌ | ✅ |

| Symbol / Pattern | Meaning | Example | Matches |
|------------------|---------|---------|----------|
| `.` | Any character except newline | `a.c` | `abc`, `a1c`, `a@c` |

| `^` | Start of string | `^Hello` | `Hello World` |
| `$` | End of string | `World$` | `Hello World` |
| `[]` | Character class (match one character from the set) | `[aeiou]` | `a`, `e`, `i` |
| `[^ ]` | Negated character class (NOT these characters) | `[^0-9]` | Any non-digit |
| `\|` | OR operator | `cat\|dog` | `cat`, `dog` |

| `(?: )` | Non-capturing group | `(?:AM\|PM)` | Groups without capturing |
| `\b` | Word boundary | `\bcat\b` | Matches whole word `cat` |
| `\B` | Not a word boundary | `\Bcat\B` | `cat` inside another word |
| `\w` | Word character | `\w+` | `Python123`, `user_007` |
| `\W` | Non-word character | `\W+` | `@#$%^` |
| `\d` | Digit | `\d` | `0-9` |
| `\D` | Non-digit | `\D+` | Letters, punctuation |
| `\s` | Whitespace | `\s+` | Spaces, tabs, newlines |
| `\S` | Non-whitespace | `\S+` | `hello`, `abc123` |
| `*` | Zero or more | `ab*` | `a`, `ab`, `abb` |
| `+` | One or more | `ab+` | `ab`, `abb`, `abbb` |
| `?` | Zero or one (optional) | `https?` | `http`, `https` |
| `{n}` | Exactly `n` times | `\d{4}` | `2025` |
| `{n,}` | At least `n` times | `a{2,}` | `aa`, `aaa`, `aaaa` |
| `{n,m}` | Between `n` and `m` times | `\d{2,4}` | `12`, `123`, `2025` |
| `*?` | Lazy zero or more | `<.*?>` | Smallest HTML tag |
| `+?` | Lazy one or more | `a+?` | Smallest possible match |
| `(?=...)` | Positive lookahead | `\d(?=kg)` | Digit followed by `kg` |
| `(?!...)` | Negative lookahead | `\d(?!kg)` | Digit not followed by `kg` |
| `(?<=...)` | Positive lookbehind | `(?<=Rs\.)\d+` | Digits after `Rs.` |
| `(?<!...)` | Negative lookbehind | `(?<!Rs\.)\d+` | Digits not preceded by `Rs.` |

# Greedy vs Lazy Quantifiers

| Greedy | Lazy | Meaning |
|---------|------|---------|
| `*` | `*?` | Zero or more |
| `+` | `+?` | One or more |
| `?` | `??` | Optional |
| `{n,m}` | `{n,m}?` | Between n and m |

Example:

```regex
<.*>
```

Matches:

```html
<div>Hello</div><p>World</p>
```

Entire string.

Whereas:

```regex
<.*?>
```

Matches:

```html
<div>
```

then


# Quick Memory Tricks

| Symbol | Remember As |
|---------|-------------|
| `.` | Any character |
| `\d` | Digit |
| `\D` | Not Digit |
| `\w` | Word character |
| `\W` | Not Word character |
| `\s` | Space |
| `\S` | Not Space |
| `+` | One or more |
| `*` | Zero or more |
| `?` | Optional |
| `{n}` | Exactly n |
| `{n,}` | At least n |
| `{n,m}` | Between n and m |
| `[]` | Choose one character |
| `[^ ]` | NOT these characters |
| `()` | Capture this |
| `(?: )` | Group, don't capture |
| `\|` | OR |
| `^` | Start |
| `$` | End |
| `\b` | Whole word boundary |
| `.` | Any character |

````python
# 30. Create a cleaned sentence ready for TF-IDF.

# step-1 converting to Lowercase
cleaned_text = text.lower()

# step-2 Remove HTMLs
cleaned_text = re.sub(r"<.*?>", " ", cleaned_text)

# step-3 Remove Urls
cleaned_text = re.sub(r"\bhttps?://\S+\b", "", cleaned_text)

# step-4 Remove emails
cleaned_text = re.sub(r"[A-Za-z0-9_%+-.]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", "", cleaned_text)

# step-5 Remove IDs/Username/password
cleaned_text = re.sub(r"username: \S+", "", cleaned_text)
cleaned_text = re.sub(r"password:\s*\S+", "", cleaned_text)
cleaned_text = re.sub(r"\bEMP_\d+_\d+\b", "", cleaned_text, flags=re.IGNORECASE)

# step-6 Remove numbers
cleaned_text = re.sub(r"\b\d+\b", "", cleaned_text)

# step-7Removing currencies if present
cleaned_text = re.sub(r"\brs\b", "", cleaned_text)

# step-8 Remove punctuation
cleaned_text = re.sub(r"[^\w\s]", "", cleaned_text)  # removes extra special characters which are inside word like you!!! becomes you

# step-9 Replace _ with " "
cleaned_text = cleaned_text.replace("_", " ")

# step-10 White space normalization
cleaned_text = re.sub(r"\s+", " ", cleaned_text)

# step-11 now my string has only trailing and ending spaces 
cleaned_text = cleaned_text.strip()

# step-12 Remove Stopwords
# As I haven't learnt nltk or spacy yet . I will make my own stop words
stop_words = {'my',"is","at", "a", "me", "and", "of", "am", "for"} # Remember to use a set as it has O(1) time complexity as compared to list having O(n).
filtered =[
    word
    for word in cleaned_text.split()
    if word not in stop_words
]


print("TF-IDF ready list is here: ",filtered)
````

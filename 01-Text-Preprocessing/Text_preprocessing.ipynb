# Text Preprocessing in NLP

---

# Overview

**Text Preprocessing** is the process of converting raw, unstructured text into a clean, standardized, and machine-readable format before applying NLP or Machine Learning algorithms.

Real-world text often contains:

- Uppercase and lowercase letters
- HTML tags
- URLs
- Emojis
- Special characters
- Punctuation
- Extra whitespaces
- Different grammatical forms of the same word
- Stopwords

Text preprocessing helps improve the quality of the input data, reduce noise, and increase the performance of NLP models.

> **Note:** There is **no universal preprocessing pipeline**. The preprocessing steps depend on the NLP task and the model being used.

---

# Why is Text Preprocessing Needed?

Consider the following sentence:

```text
"The CHILDREN were RUNNING quickly towards the beautiful gardens! Visit https://example.com 😊"
```

Although humans understand it easily, a computer simply sees it as a sequence of characters.

After preprocessing, the sentence becomes much cleaner and standardized:

```text
child run quickly towards beautiful garden visit
```

This standardized text is much easier for classical NLP algorithms to process.

---

# Step 1: Original Text

```python
text = "The CHILDREN were RUNNING quickly towards the beautiful gardens! Visit https://example.com 😊"

print(text)
```

### Output

```text
The CHILDREN were RUNNING quickly towards the beautiful gardens! Visit https://example.com 😊
```

---

# Step 2: Lowercasing

Convert all characters into lowercase.

```python
text = text.lower()

print(text)
```

### Output

```text
the children were running quickly towards the beautiful gardens! visit https://example.com 😊
```

### Why?

- Removes case differences.
- Treats **Dog**, **DOG**, and **dog** as the same word.

---

# Step 3: Remove URLs

```python
import re

text = re.sub(r'https?://\S+', '', text)

print(text)
```

### Output

```text
the children were running quickly towards the beautiful gardens! visit
```

### Why?

URLs usually do not contribute useful information for many NLP tasks.

---

# Step 4: Remove Special Characters and Emojis

```python
text = re.sub(r'[^a-zA-Z\s]', '', text)

print(text)
```

### Output

```text
the children were running quickly towards the beautiful gardens visit
```

### Why?

Special characters and emojis often introduce unnecessary noise in classical NLP tasks.

---

# Step 5: Remove Extra Whitespaces

```python
text = re.sub(r'\s+', ' ', text).strip()

print(text)
```

### Output

```text
the children were running quickly towards the beautiful gardens visit
```

### Why?

Creates a clean and standardized sentence.

---

# Step 6: Tokenization

Split the sentence into individual tokens.

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)

print(tokens)
```

### Output

```python
[
'the',
'children',
'were',
'running',
'quickly',
'towards',
'the',
'beautiful',
'gardens',
'visit'
]
```

### Why?

Most NLP algorithms operate on **tokens**, not raw text.

---

# Step 7: POS Tagging

Assign a grammatical role to every token.

```python
from nltk import pos_tag

pos_tags = pos_tag(tokens)

print(pos_tags)
```

### Output

```python
[
('the', 'DT'),
('children', 'NNS'),
('were', 'VBD'),
('running', 'VBG'),
('quickly', 'RB'),
('towards', 'IN'),
('the', 'DT'),
('beautiful', 'JJ'),
('gardens', 'NNS'),
('visit', 'NN')
]
```

### Why?

The same word can have different meanings depending on its grammatical role.

Example:

```text
Book the tickets.
```

↓

`Book` → Verb

```text
This book is interesting.
```

↓

`book` → Noun

POS tagging helps the lemmatizer determine the correct base form.

---

# Step 8: Convert Penn Treebank Tags to WordNet Tags

```python
from nltk.corpus import wordnet

def get_wordnet_pos(tag):

    if tag.startswith("J"):
        return wordnet.ADJ

    elif tag.startswith("V"):
        return wordnet.VERB

    elif tag.startswith("N"):
        return wordnet.NOUN

    elif tag.startswith("R"):
        return wordnet.ADV

    else:
        return wordnet.NOUN
```

### Why?

`pos_tag()` returns **Penn Treebank tags**, while `WordNetLemmatizer` expects **WordNet POS tags**.

---

# Step 9: Lemmatization

Convert words into their dictionary form.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_tokens = []

for word, tag in pos_tags:

    wn_tag = get_wordnet_pos(tag)

    lemma = lemmatizer.lemmatize(word, pos=wn_tag)

    lemmatized_tokens.append(lemma)

print(lemmatized_tokens)
```

### Output

```python
[
'the',
'child',
'be',
'run',
'quickly',
'towards',
'the',
'beautiful',
'garden',
'visit'
]
```

Notice:

```text
children → child
were → be
running → run
gardens → garden
```

---

# Step 10: Stopword Removal (Optional)

Remove common words that usually carry little semantic meaning.

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

filtered_tokens = [
    word
    for word in lemmatized_tokens
    if word not in stop_words
]

print(filtered_tokens)
```

### Output

```python
[
'child',
'run',
'quickly',
'towards',
'beautiful',
'garden',
'visit'
]
```

### Why After Lemmatization?

Lemmatization relies on **POS tagging**, and POS tagging performs better when it has the complete sentence.

Removing stopwords before POS tagging removes grammatical context and may reduce tagging accuracy.

---

# Step 11: Join Tokens (Optional)

```python
processed_text = " ".join(filtered_tokens)

print(processed_text)
```

### Output

```text
child run quickly towards beautiful garden visit
```

This cleaned sentence is now ready for feature extraction techniques such as:

- Bag of Words (BoW)
- TF-IDF
- Word Embeddings

# Classical NLP vs Modern NLP

## Classical NLP

A preprocessing pipeline is commonly used.

```text
Raw Text
      │
      ▼
Cleaning
      │
      ▼
Tokenization
      │
      ▼
POS Tagging
      │
      ▼
Lemmatization
      │
      ▼
Stopword Removal
      │
      ▼
Bag of Words / TF-IDF
```

---

## Modern NLP (Transformers)

Modern Transformer models generally perform **minimal preprocessing**.

```text
Raw Text
      │
      ▼
Subword Tokenization
      │
      ▼
Transformer Model
```

Typically, they **do not**:

- Remove stopwords
- Perform stemming
- Perform lemmatization

The model learns these linguistic relationships directly from raw text.

# Summary

| Step | Purpose |
|------|---------|
| Lowercasing | Standardize text by removing case differences |
| Remove URLs | Remove web links |
| Remove Special Characters | Remove punctuation, emojis, and symbols |
| Remove Extra Whitespaces | Clean spacing |
| Tokenization | Split text into meaningful tokens |
| POS Tagging | Assign grammatical roles to tokens |
| Lemmatization | Convert words to their dictionary forms |
| Stopword Removal | Remove common words if required |
| Join Tokens | Create cleaned text for feature extraction |

---

# Key Takeaways

- **Text preprocessing** prepares raw text for NLP and Machine Learning models.
- There is **no fixed preprocessing pipeline**; the required steps depend on the task.
- POS tagging should generally be performed **before lemmatization** because the lemmatizer needs grammatical information.
- Stopword removal is usually performed **after lemmatization** in POS-aware pipelines.
- Classical NLP relies heavily on preprocessing, whereas modern Transformer-based models typically use minimal preprocessing.
- A well-designed preprocessing pipeline improves consistency, reduces noise, and prepares text for feature extraction methods such as **Bag of Words** and **TF-IDF**.

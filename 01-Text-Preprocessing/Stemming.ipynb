# Stemming

---

# Overview

**Stemming** is the process of reducing a word to its **stem (root)** by removing prefixes or suffixes using predefined rules.

The main objective of stemming is to reduce different forms of a word to a common representation so that they are treated as the same word during text analysis.

Unlike lemmatization, **stemming does not consider the meaning or grammatical role of a word**.

> **Note:** The resulting stem is **not guaranteed to be a valid English dictionary word.**

---

# Why is Stemming Needed?

Consider the following words:

```text
play
playing
played
plays
```

Although these words represent the same concept, a computer treats them as four completely different words.

Without stemming:

```python
["play", "playing", "played", "plays"]
```

After stemming:

```python
["play", "play", "play", "play"]
```

Now the vocabulary size becomes smaller, making NLP models more efficient.

---

# Example

Original Sentence:

```text
The boys were playing football and later played cricket.
```

After Tokenization:

```python
['The', 'boys', 'were', 'playing', 'football', 'and', 'later', 'played', 'cricket']
```

After Stemming:

```python
['The', 'boy', 'were', 'play', 'football', 'and', 'later', 'play', 'cricket']
```

Notice that **playing** and **played** are reduced to the same stem.

---

# How Stemming Works

A stemmer simply applies predefined rules to remove common prefixes or suffixes.

For example:

```text
playing
```

↓

Remove **ing**

↓

```text
play
```

Another example:

```text
studies
```

↓

Remove **es**

↓

```text
studi
```

Notice that **studi** is **not** a valid English word.

This is acceptable because the purpose of stemming is **normalization**, not producing meaningful words.

---

# Advantages of Stemming

- Reduces vocabulary size
- Faster than lemmatization
- Simple and computationally efficient
- Improves performance of classical NLP models
- Useful for search engines and information retrieval

---

# Disadvantages of Stemming

- Stem may not be a valid English word
- Does not understand grammar
- Does not consider context
- Can over-stem or under-stem words
- Less accurate than lemmatization

Example:

```text
studies
```

↓

```text
studi
```

Example:

```text
university
```

↓

```text
univers
```

These are stems but not valid dictionary words.

---

# Popular Stemming Algorithms

NLTK provides four commonly used stemmers:

1. Porter Stemmer
2. Snowball Stemmer
3. Lancaster Stemmer
4. Regexp Stemmer

---

# 1. Porter Stemmer

The **Porter Stemmer** is the most widely used stemming algorithm.

It was developed by **Martin Porter** in 1980 and provides a good balance between speed and accuracy.

## Import

```python
from nltk.stem import PorterStemmer
```

## Create Object

```python
ps = PorterStemmer()
```

## Example

```python
words = [
    "playing",
    "quickly",
    "jumps",
    "studies",
    "happiness",
    "better",
    "played",
    "foxes",
    "University"
]

stemmed_words = [ps.stem(word) for word in words]

print(stemmed_words)
```

Output

```python
[
'play',
'quickli',
'jump',
'studi',
'happi',
'better',
'play',
'fox',
'univers'
]
```

### Observation

- Very fast
- Balanced stemming
- Stem may not be a valid English word

Examples:

```text
quickly → quickli
studies → studi
happiness → happi
```

---

# 3. Lancaster Stemmer

The **Lancaster Stemmer** is the most aggressive stemmer available in NLTK.

It removes suffixes more aggressively than Porter and Snowball.

## Import

```python
from nltk.stem import LancasterStemmer
```

## Create Object

```python
stemmer = LancasterStemmer()
```

## Example

```python
words = [
    "playing",
    "quickly",
    "jumps",
    "studies",
    "happiness",
    "better",
    "played",
    "foxes",
    "University"
]

stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

Output

```python
[
'play',
'quick',
'jump',
'study',
'happy',
'bet',
'play',
'fox',
'univers'
]
```

### Observation

Lancaster performs aggressive stemming.

Examples:

```text
better → bet
```

```text
happiness → happy
```

```text
studies → study
```

Although it often produces valid words, it can also over-stem words.

---

# 4. Regexp Stemmer

The **Regexp Stemmer** allows users to define their own stemming rules using regular expressions.

It provides complete control over which suffixes should be removed.

## Import

```python
from nltk.stem import RegexpStemmer
```

## Create Object

```python
stemmer = RegexpStemmer(r"ing$")
```

## Example

```python
words = [
    "playing",
    "played",
    "running",
    "runs"
]

stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

Output

```python
[
'play',
'played',
'runn',
'runs'
]
```

### Observation

The regex

```python
r"ing$"
```

means:

- Remove **ing**
- Only if it appears at the **end of the word**

Therefore:

```text
playing → play
running → runn
played → played
runs → runs
```

# When Should You Use Stemming?

Use stemming when:

- Speed is important
- Exact dictionary words are not required
- Working with classical machine learning models
- Performing document retrieval or search

Avoid stemming when:

- Grammar and meaning are important
- Building chatbots
- Machine Translation
- Text Summarization
- Question Answering
- Large Language Models (LLMs)

---

# Stemming vs Lemmatization

| Feature | Stemming | Lemmatization |
|----------|-----------|---------------|
| Uses Rules | ✅ Yes | ❌ No |
| Uses Dictionary | ❌ No | ✅ Yes |
| Uses Grammar (POS) | ❌ No | ✅ Yes |
| Produces Valid Words | ❌ Not Always | ✅ Yes |
| Speed | Fast | Slower |
| Accuracy | Lower | Higher |

Example:

| Original Word | Stem | Lemma |
|---------------|------|--------|
| playing | play | play |
| studies | studi | study |
| happiness | happi | happiness |
| better | better / bet | good |

# Key Takeaways

- Stemming reduces words to their **stem** using predefined rules.
- The resulting stem **may not be a valid English word**.
- It is much faster than lemmatization because it does not use dictionaries or grammatical information.
- Porter Stemmer is the most commonly used stemmer in classical NLP.
- Snowball Stemmer is an improved version of Porter.
- Lancaster Stemmer is more aggressive and may over-stem words.
- Regexp Stemmer allows complete customization using regular expressions.
- Stemming is mainly used in **classical NLP**, while **lemmatization** is preferred when preserving the actual meaning of words is important.

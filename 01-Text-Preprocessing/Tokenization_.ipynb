````python
text = """
Hello!! what's happening.
I'm Roshan, I just love nlp. OOps!! I forgot to bring my laptop.
Can't I work on yours.
"""
````

### Sentence tokenizer: sent_tokenize

````python
import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
````

````python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
print(sentences)
````

### Sentence Tokenization: PunktSentenceTokenizer

````python
# PunktSentenceTokenizer is a class and we make object of that class and then call its function.
import nltk
from nltk.tokenize import PunktSentenceTokenizer

nltk.download("punkt")
nltk.download("punkt_tab")

tokenizer = PunktSentenceTokenizer()
sentences = tokenizer.tokenize(text)

print(sentences)
````

### Word Tokenization: word_tokenize

````python
from nltk.tokenize import word_tokenize

words = word_tokenize(text)
print(words)
````

### Word Tokenization: TreebankWordTokenizer

````python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
words = tokenizer.tokenize(text)
print(words)
````

### Word Tokenization: WordPunktTokenizer

````python
from nltk.tokenize import WordPunctTokenizer

tokenizer = WordPunctTokenizer()
words= tokenizer.tokenize(text)
print(words)
````

### Word Tokenization: RegexpTokenizer

````python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
words = tokenizer.tokenize(text)

print(words)
````

### Word Tokenization: spaCy Tokenizer

````python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(text)
words = []
for token in doc:
    words.append(token.text)

print(words)
````

# Tokenization

# Overview

**Tokenization** is the process of breaking a piece of text into smaller meaningful units called **tokens**.

These tokens are the basic units on which almost every NLP algorithm operates.

A token can be:

- A sentence
- A word
- A character
- A subword

**Example:**

```text
Input:
I love learning NLP.
```

After **Word Tokenization**:

```python
["I", "love", "learning", "NLP", "."]
```

Notice that punctuation can also become a token.

---

# Why is Tokenization Needed?

Computers cannot understand raw text.

For a computer,

```text
I love NLP.
```

is simply one long string.

After tokenization:

```python
["I", "love", "NLP", "."]
```

each token can now be processed individually.

Tokenization is required for:

- Stopword Removal
- Stemming
- Lemmatization
- POS Tagging
- Named Entity Recognition (NER)
- Bag of Words (BoW)
- TF-IDF
- Word Embeddings
- Transformers (BERT, GPT, Llama, etc.)

Without tokenization, almost every NLP preprocessing technique becomes difficult or impossible.

---

# Types of Tokenization

Tokenization can be classified based on **what the text is split into**.

## 1. Sentence Tokenization

Sentence tokenization splits a paragraph into individual sentences.

### Example

```text
Hello! My name is Alex. I love NLP.
```

↓

```python
[
    "Hello!",
    "My name is Alex.",
    "I love NLP."
]
```

### Applications

- Document Summarization
- Machine Translation
- Question Answering
- Sentence Classification
- Information Retrieval

---

## 2. Word Tokenization

Word tokenization splits each sentence into words (or word-like units).

### Example

```text
I love NLP.
```

↓

```python
["I", "love", "NLP", "."]
```

Notice that punctuation may also become a token.

### Applications

This is the most commonly used type of tokenization.

Used in:

- Bag of Words
- TF-IDF
- Sentiment Analysis
- Spam Detection
- Text Classification
- Topic Modelling

---

## 3. Character Tokenization

Each individual character becomes a token.

### Example

```text
NLP
```

↓

```python
['N', 'L', 'P']
```

### Applications

- OCR
- Character-level Language Models
- Spell Correction
- DNA Sequence Analysis

---

## 4. Subword Tokenization

Instead of splitting into complete words, the tokenizer splits words into meaningful subword units.

### Example

```text
playing
```

↓

```text
play
ing
```

Another example

```text
unbelievable
```

↓

```text
un
believ
able
```

### Why?

Subword tokenization solves the **Out-of-Vocabulary (OOV)** problem.

Even if a model has never seen the word:

```text
electroencephalography
```

it can still process it by splitting it into smaller known subwords.

### Applications

Used in modern Transformer models such as:

- GPT
- BERT
- Llama
- T5

---

# Approaches to Tokenization

Tokenization can also be classified based on **how the splitting is performed**.

---

## 1. Whitespace Tokenization

The simplest approach.

Text is split wherever whitespace occurs.

### Example

```python
text = "I love NLP"

text.split()
```

Output

```python
["I", "love", "NLP"]
```

### Advantages

- Very simple
- Very fast
- Built into Python

### Disadvantages

Cannot correctly handle:

```text
I'm
NLP.
Hello,
```

---

## 2. Rule-Based Tokenization

Uses predefined linguistic rules.

Examples of rules:

- Separate punctuation
- Handle contractions
- Handle abbreviations
- Handle quotation marks

### Example

```text
I'm learning NLP.
```

↓

```python
["I", "'m", "learning", "NLP", "."]
```

### Advantages

- More accurate than whitespace tokenization
- Handles punctuation correctly
- Handles contractions

### Examples

- TreebankWordTokenizer
- word_tokenize()
- spaCy Tokenizer

---

## 3. Statistical Tokenization

Instead of manually defining rules, statistical tokenizers learn token boundaries from a large corpus.

Example:

```text
Dr. Smith
```

A statistical tokenizer learns that

```text
Dr.
```

usually does **not** indicate the end of a sentence.

### Example

Punkt Sentence Tokenizer

Used internally by:

```python
sent_tokenize()
```

### Advantages

- Learns language patterns automatically
- Better sentence boundary detection

---

## 4. Byte Pair Encoding (BPE)

BPE is a **subword tokenization algorithm**.

Instead of treating unknown words as completely new words, BPE splits them into frequent subword units.

Example

```text
playing
```

↓

```text
play
ing
```

### Advantages

- Smaller vocabulary
- Handles unseen words
- Very efficient

### Applications

Used by many GPT-family models.

---

## 5. Transformer-Based Tokenization

Modern Transformer models use **subword tokenization** rather than word tokenization.

Different Transformer models use different algorithms.

Examples:

- Byte Pair Encoding (BPE)
- WordPiece
- SentencePiece
- Unigram

These tokenizers also convert each token into a unique integer called a **Token ID**.

Example

```text
I love NLP
```

↓

```python
["I", "love", "NLP"]
```

↓

```python
[72, 431, 9812]
```

These numerical IDs are what the neural network actually processes.

# Types of Word Tokenizers

The following tokenizers are commonly available in Python and NLTK.

---

## 1. Python `split()`

The simplest tokenizer.

Splits text using whitespace.

```python
text.split()
```

### Advantages

- Very fast
- No external library required

### Disadvantages

- Does not separate punctuation
- Does not handle contractions
- Not suitable for real NLP tasks

---

## 2. `word_tokenize()`

General-purpose tokenizer provided by NLTK.

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
```

### Features

- Handles punctuation
- Handles contractions
- Handles abbreviations better than `split()`

Most commonly used tokenizer in classical NLP.

---

## 3. `TreebankWordTokenizer`

Rule-based tokenizer following the Penn Treebank tokenization rules.

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

tokens = tokenizer.tokenize(text)
```

### Features

- Separates contractions
- Uses rule-based tokenization
- Context-aware handling of punctuation

---

## 4. `WordPunctTokenizer`

Separates words from punctuation.

```python
from nltk.tokenize import WordPunctTokenizer

tokenizer = WordPunctTokenizer()

tokens = tokenizer.tokenize(text)
```

### Features

- Splits words and punctuation separately
- Consecutive punctuation is usually grouped together

Example

```text
Hello!!!
```

↓

```python
["Hello", "!!!"]
```

---

## 5. `RegexpTokenizer`

Allows custom tokenization using regular expressions.

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

tokens = tokenizer.tokenize(text)
```

### Features

- User defines tokenization rules
- Highly customizable
- Useful for domain-specific preprocessing

---

## 6. spaCy Tokenizer

Tokenizer provided by spaCy.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

tokens = [token.text for token in doc]
```

### Features

- Rule-based + language-aware
- Integrates with the complete spaCy pipeline
- Every token is an object containing linguistic information

Example:

```python
token.text
token.lemma_
token.pos_
token.dep_
token.is_stop
token.is_alpha
```

# Key Takeaways

- Tokenization is the foundation of every NLP pipeline.
- It converts raw text into smaller meaningful units called **tokens**.
- Tokenization can be performed at the sentence, word, character, or subword level.
- Different tokenization approaches exist depending on the application.
- Classical NLP primarily relies on **word tokenization**, while modern Transformer models rely on **subword tokenization**.
- Choosing the right tokenizer depends on the NLP task, the language, and the downstream model.

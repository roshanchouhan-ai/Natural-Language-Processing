# Part-of-Speech (POS) Tagging

---

# Overview

**Part-of-Speech (POS) Tagging** is the process of assigning a **grammatical category** (also called a **POS tag**) to every word in a sentence.

Each word is tagged according to its role in the sentence, such as:

- Noun
- Verb
- Adjective
- Adverb
- Pronoun
- Preposition
- Conjunction
- Determiner
- Interjection

POS tagging helps computers understand the grammatical structure of a sentence.

---

# Why is POS Tagging Needed?

Many English words can have **different meanings depending on how they are used in a sentence**.

For example:

### Sentence 1

```text
This book is very interesting.
```

Here,

```text
book
```

is a **Noun**.

---

### Sentence 2

```text
Book the tickets before tomorrow.
```

Here,

```text
book
```

is a **Verb**.

Although both words have the same spelling, their meanings are different.

POS tagging helps the computer distinguish between them.

---

# Another Example

```text
Running is healthy.
```

Here,

```text
Running
```

is a **Noun**.

---

```text
He is running very fast.
```

Here,

```text
running
```

is a **Verb**.

Again, the same word performs different grammatical roles.

---

# Why is POS Tagging Important?

POS tagging is an important preprocessing step because many NLP tasks depend on knowing the grammatical role of each word.

It is used in:

- Lemmatization
- Named Entity Recognition (NER)
- Dependency Parsing
- Machine Translation
- Question Answering
- Text Summarization
- Information Extraction
- Grammar Checking
- Speech Recognition

# Common POS Tags

NLTK follows the **Penn Treebank Tagset**.

Some of the most common tags are:

| Tag | Meaning | Example |
|------|----------|----------|
| NN | Singular Noun | cat |
| NNS | Plural Noun | cats |
| NNP | Proper Noun | India |
| NNPS | Proper Noun (Plural) | Himalayas |
| VB | Verb (Base Form) | play |
| VBD | Verb (Past Tense) | played |
| VBG | Verb (Gerund / Present Participle) | playing |
| VBN | Verb (Past Participle) | eaten |
| VBP | Verb (Present Tense) | play |
| VBZ | Verb (3rd Person Singular) | plays |
| JJ | Adjective | beautiful |
| JJR | Comparative Adjective | better |
| JJS | Superlative Adjective | best |
| RB | Adverb | quickly |
| RBR | Comparative Adverb | faster |
| RBS | Superlative Adverb | fastest |
| PRP | Personal Pronoun | he, she |
| DT | Determiner | the, a |
| IN | Preposition | in, on, under |
| CC | Coordinating Conjunction | and, but |
| CD | Cardinal Number | 10 |
| UH | Interjection | wow |

# POS Tagging and Lemmatization

Lemmatization works much better when it knows the grammatical role of a word.

Example:

```python

```

Output

```text
playing
```

The lemmatizer assumes that the word is a **noun** by default.

---

If we specify that it is a **verb**:

```python

```

Output

```text

```

This demonstrates why POS tagging is often performed before lemmatization.

# Converting NLTK POS Tags to WordNet POS Tags

NLTK's `pos_tag()` returns **Penn Treebank tags**, while the `WordNetLemmatizer` expects **WordNet POS constants**.

The mapping is:

| NLTK Tag Starts With | WordNet POS |
|----------------------|-------------|
| J | wordnet.ADJ |
| V | wordnet.VERB |
| N | wordnet.NOUN |
| R | wordnet.ADV |
| Others | wordnet.NOUN (default) |

---

## Helper Function

```python
from nltk.corpus import wordnet

def get_wordnet_pos(tag):

    if tag.startswith("J"):
        return wordnet.ADJ

    elif tag.startswith("V"):
        return wordnet.VERB

    elif tag.startswith("N"):
        return wordnet.NOUN

    elif tag.startswith("R"):
        return wordnet.ADV

    else:
        return wordnet.NOUN
```

# Complete Example

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

text = "The children were running quickly towards the beautiful gardens."

tokens = word_tokenize(text)

tags = pos_tag(tokens)

lemmatizer = WordNetLemmatizer()

lemmatized_words = []

for word, tag in tags:
    wn_tag = get_wordnet_pos(tag)
    lemma = lemmatizer.lemmatize(word, pos=wn_tag)
    lemmatized_words.append(lemma)

print(lemmatized_words)
```

Output

```python
[
'The',
'child',
'be',
'run',
'quickly',
'towards',
'the',
'beautiful',
'garden',
'.'
]
```

Notice:

```text
children → child
were → be
running → run
gardens → garden
```

This would not be possible without proper POS tagging.

# Advantages of POS Tagging

- Improves lemmatization accuracy
- Helps understand sentence structure
- Resolves ambiguity in words
- Improves many downstream NLP tasks
- Essential for syntactic analysis

---

# Limitations

- Accuracy depends on the quality of the POS tagger
- Ambiguous sentences can still be tagged incorrectly
- Different languages require different taggers
- Context is sometimes difficult to determine

# Key Takeaways

- POS Tagging identifies the grammatical role of every word in a sentence.
- The same word can have different POS tags depending on its context.
- NLTK uses the **Penn Treebank Tagset** for POS tagging.
- `pos_tag()` returns a list of `(word, tag)` tuples.
- Lemmatization becomes much more accurate when supplied with the correct POS tag.
- Before using `WordNetLemmatizer`, Penn Treebank tags should be mapped to the corresponding WordNet POS constants.
- POS Tagging is a fundamental building block for many advanced NLP tasks such as Named Entity Recognition, Dependency Parsing, Machine Translation, and Question Answering.
